# Random Forest Regression

## Importing the libraries

```
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

## Importing the dataset

```
dataset = pd.read_csv('Ice_Cream_Sales.csv')
X = dataset.iloc[:, :-1].values
y = dataset.iloc[:, -1].values
```

```
dataset.head()
```

```
dataset.info()
```

## Training the Random Forest Regression model on the whole dataset



```
from sklearn.ensemble import RandomForestRegressor
regressor = RandomForestRegressor(n_estimators = 10, random_state = 0)
regressor.fit(X, y)
```

## Predicting a new result

```
regressor.predict([[6.5]])
```

## Visualising the Random Forest Regression results (higher resolution)

```
# Create a range of values from the minimum to maximum of X with small step size (0.01)
X_grid = np.arange(min(X), max(X), 0.01)

# Reshape X_grid to be a 2D array, required for prediction
X_grid = X_grid.reshape((len(X_grid), 1))

# Scatter plot of actual data points (red dots)
plt.scatter(X, y, color='red')

# Plot the model’s predictions as a smooth blue curve
plt.plot(X_grid, regressor.predict(X_grid), color='blue')

# Set the title of the plot
plt.title('Truth or Bluff (Random Forest Regression)')

# Label the x-axis
plt.xlabel('Temperature')

# Label the y-axis
plt.ylabel('Sales (Units)')

# Display the plot
plt.show()
```
