```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.api as sm

# Upload the dataset
from google.colab import files
uploaded = files.upload()

# Load the dataset
filename = list(uploaded.keys())[0]
data = pd.read_excel(filename)

# Display basic info
print("Dataset Overview:")
print(data.head())
print(data.info())

# Data Cleaning and Preparation
# Checking for missing values
print("Missing Values:\n", data.isnull().sum())

# Assuming 'Marketing_Spend' and 'Revenue' are columns in the dataset
data = data.dropna()  # Remove rows with missing values if necessary

# Define features and target variable
X = data['Marketing_Spend'].values.reshape(-1, 1)  # Feature column reshaped to 2D
y = data['Monthly_Revenue'].values.reshape(-1, 1)  # Target column

# Create a DataFrame
df = pd.DataFrame(data)

# Calculate xy, x^2, and y^2
df['xy'] = df['Marketing_Spend'] * df['Monthly_Revenue']
df['x_squared'] = df['Marketing_Spend'] ** 2
df['y_squared'] = df['Monthly_Revenue'] ** 2

# Display the DataFrame
df
```

```python
# Calculate necessary values
n = len(df)
sum_x = df['Marketing_Spend'].sum()
sum_y = df['Monthly_Revenue'].sum()
sum_xy = (df['Marketing_Spend'] * df['Monthly_Revenue']).sum()
sum_x_squared = (df['Marketing_Spend'] ** 2).sum()
sum_y_squared = (df['Monthly_Revenue'] ** 2).sum()

# Calculate Pearson correlation coefficient
r_numerator = n * sum_xy - sum_x * sum_y
r_denominator = np.sqrt((n * sum_x_squared - sum_x**2) * (n * sum_y_squared - sum_y**2))
r = r_numerator / r_denominator
print("Pearson correlation coefficient:", r)
```
